```python
from ultralytics import YOLO
import cv2
import cvzone

```

```python
cam = cv2.VideoCapture(r"C:\Users\soham\OneDrive\Desktop\YOLO\ppe-2-1.mp4")
cam.set(3,640)
cam.set(4,480)

className = ["Hardhat","Mask","NO-hardhat","NO-mask","NO-Safety Vest","Person","Safety Cone","safety Vest","Machinery","Vehicle"]

model = YOLO("best.pt")

while True:
    sucess, img = cam.read()
    if not sucess:
        break

    result = model(img,stream=True,conf = 0.5)
    for r in result:

        for box in r.boxes:
            x1,y1,x2,y2 = map(int,box.xyxy[0])
            print(x1,y1,x2,y2)

            conf = round(float(box.conf[0]),2)
            print(conf)

            cls = int(box.cls[0])

            currentClass = className[cls]
            print(currentClass)

            if currentClass == "NO-hardhat" or currentClass == "NO-mask" or currentClass == "NO-Safety Vest":
                myColor = (0,0,255)
        
            else:
                myColor = (0,255,0)
            
            cv2.rectangle(img,(x1,y1),(x2,y2),myColor,2)
            cvzone.putTextRect(img,f'{currentClass} {conf}',(max(0,x1),max(35,y1)),scale= 1,thickness=1)



    cv2.imshow("Image",img)
    cv2.waitKey(1)
```

```python
cam.release()
cv2.destroyAllWindows()
```
